```python
#pivote table
df=pd.DataFrame({'A':['a','b','c','d']*3,
                 'B':['A','B','C',]*4,
                 'C':['P','P','P','Q','Q','Q']*2,
                 'D':np.random.randn(12),
                 'E':np.random.randn(12)})
df
```
